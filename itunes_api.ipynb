```python
import requests
import json
```

```python
# itunes serarch api
base_url = 'https://itunes.apple.com/search'
url = base_url+"?term=the+beatles&country=us"
requests.get(url)
```

```python
r= requests.get(base_url,params = {"term":"the beatles","country":"us"})
r.status_code
```

```python
info = r.json()
print(json.dumps(info,indent=5))
```

```python
# check dict keys from output
info.keys()
```

```python
# inspect result object items
json.dumps(info['results'][0], indent=4)
```

```python
info["resultCount"]
```

```python
# set result count limit
r= requests.get(base_url,params = {"term":"the beatles","country":"us","limit":400})
r.ok
```
